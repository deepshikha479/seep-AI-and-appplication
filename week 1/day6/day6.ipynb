#Filtering Data in a DataFrame

```python
#Creating a dictionary
data ={
    'Name':['Array','Ramesh','Monkia'],
    'Class':[7,8,9],
    'Roll' :[1,2,3],
    'Address': ['Pokhara','Butwal','Jhapa']
}
#Turning dictionary into a dataFrame
df= pd.DataFrame(data)

#Filtering
print(df[df["Class"] > 7])



```

#basic plotting with matplotlib

```python
import matplotlib.pyplot as plt
#Defining Data for basic plotting
x= [1,2,3,4,5]
y=[2,4,6,8,10]

plt.plot(x,y,marker="o")
plt.title("Plotting Example")
plt.xlable("X-axis")
plt.ylable("Y-axis")
plt.grid(True)
plt.savefig("my_frist_plot.png")
plt.show()
```

#visuaalizing a data a dataframe with matplotlib

```python
data = {
    "Name": ["Aarav", "Ramesh", "Monkia"],
    "Class": [7, 8, 9],
}

df = pd.DataFrame(data)

plt.bar(df["Name"], df["Class"])
plt.title("Class Distribution")
plt.xlabel("Name")
plt.ylabel("Class")
plt.grid(True)
plt.show()
```
