##1. Measure of central tenency

```python
import pandas as pd
data = pd.Series([34,67,39,69,98,111])

mean =data.mean()
median = data.median()
mode = data.mode()
print("the mean of the data is:",mean)
print("the median of the data is:",median)
print("the mode of the data is:",mode)
```

##2. MEASURE OF VARIABILITY

```python
data = pd.Series([34,67,39,69,98,111])

range =data.max() - data.min()
print("the range of the data is:",range)
```

##VARIANCE AND SD


```python
data = pd.Series([34,67,39,69,98,111])
variance = data.var()
standard_deviation = data.std()
print(variance)
print(standard_deviation)
```

##VISUALIZING DATA WITH MATPLOTLIB

```python
import matplotlib.pyplot as plt
data = pd.DataFrame({
    "Scores": [-100,25,78,45,14,78,45,34,66,100,400,600,700]
})

data.boxplot(column="Scores")
plt.show()
```
